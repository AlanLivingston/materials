## <span class="c7 c17">Problem 3</span>

<span class="c5 c0">Nowadays we take word completion for granted. Our phones, text editors, and word processing programs all give us suggestions for how to complete words as we type based on the letters typed so far. These hints help speed up user input and eliminate common typographical mistakes (but can also be frustrating when the tool insists on completing a word that you don’t want completed).</span>

<span class="c0">You will implement two functions that such tools might use to provide command completion. The first function,</span> <span class="c3">fill_competions</span><span class="c0">, will construct a dictionary designed to permit easy calculation of possible word completions. A problem for any such function is what vocabulary, or set of words, to allow completion on. Because the vocabulary you want may depend on the domain a tool is used in, you will provide</span> <span class="c3">fill_competions</span><span class="c0">with a representative sample of documents from which it will build the completions dictionary. The second function,</span> <span class="c3">find_completions</span><span class="c5 c0">, will return the set of possible completions for a start of any word in the vocabulary (or the empty set if there are none). In addition to these two functions, you will implement a simple main program to use for testing your functions.</span>

## <span class="c17 c7">Specifications</span>

*   <span class="c3">fill_completions(fd)</span><span class="c0"> returns a dictionary. This function takes as input an opened file. It loops through each line in the file, splitting the lines into individual words (separated by whitespace) and builds a dictionary:</span>

*   <span class="c0">The keys of the dictionary are tuples of the form</span> <span class="c3">(n, l)</span><span class="c0">for a non-negative integer</span> <span class="c3">n</span><span class="c0">and a lowercase letter</span> <span class="c3">l</span><span class="c0">.</span>
*   <span class="c0">The value associated with key</span> <span class="c3">(n, l)</span><span class="c0">is the set of words in the file that contain the letter</span> <span class="c3">l</span><span class="c0">at position</span> <span class="c3">n</span><span class="c0">. For simplicity, all vocabulary words are converted to lower case. For example, if the file contains the word "Python" and</span> <span class="c3">c_dict</span><span class="c0">is the returned dictionary, then the sets</span> <span class="c3">c_dict[0, "p"]</span><span class="c0">,</span> <span class="c3">c_dict[1, "y"]</span><span class="c0">,</span> <span class="c3">c_dict[2, "t"]</span><span class="c0">,</span> <span class="c3">c_dict[3, "h"]</span><span class="c0">,</span> <span class="c3">c_dict[4, "o"]</span><span class="c0">, and</span> <span class="c3">c_dict[5, "n"]</span><span class="c0">all contain the word</span> <span class="c3">"python"</span><span class="c0">.</span>
*   <span class="c0">Words are stripped of leading and trailing punctuation.</span>
*   <span class="c0">Words containing non-alphabetic characters are ignored, as are words of length 1 (since there is no reason to complete the latter).</span>

*   <span class="c3">find_completions(prefix, c_dict)</span><span class="c0"> returns a set of strings. This function takes a prefix of a vocabulary word and a completions dictionary of the form described above. It returns the set of vocabulary words in the completions dictionary, if any, that complete the prefix. It the prefix cannot be completed to any vocabulary words, the function returns an empty set.</span>
*   <span class="c3">main()</span><span class="c0">, the test driver:</span>

*   <span class="c0">Opens the file named</span> <span class="c3">articles.txt</span><span class="c0">. This file contains the text of recent articles pulled from BBC.</span>
*   <span class="c0">Calls</span> <span class="c3">fill_competions</span><span class="c0"> to fill out a completions dictionary using this file.</span>
*   <span class="c0">Repeatedly prompts the user for a prefix to complete.</span>
*   <span class="c0">Prints each word from the set of words that can complete the given prefix (one per line). If no completions are possible, it should just print "No completions".</span>
*   <span class="c0">Quit if the user enters the word "<quit>".</span>
